##  Find the most prolific writers and analyze their contributions to movies in terms of ratings and genre.

```python
writer_counts = df['writer'].value_counts()
```

```python
writer_counts
```

```python
prolific_writers = writer_counts.head(10) 
```

```python
prolific_writers
```

```python
writer_analysis_df = df[df['writer'].isin(prolific_writers.index)]
```

```python
writer_analysis_df
```

```python
writer_analysis_df.groupby('writer').mean()
```

```python
writer_analysis_df['profit'] = writer_analysis_df['gross'] - writer_analysis_df['budget']

writer_analysis_df
```

```python
writer_profits = writer_analysis_df.groupby('writer')['profit'].sum()
```

```python
writer_profits 
writer_profits_df = writer_profits.reset_index()
writer_profits_df.columns = ['writer', 'profit']

# Sort the writers by total profit in descending order
#writer_profits = writer_profits.sort_values(ascending=False)

# Display the top writers by profit
#writer_profits
```

```python
writer_profits_df.sort_values(by='profit',ascending=False)
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(data=writer_profits_df,y='writer',x='profit')
```
